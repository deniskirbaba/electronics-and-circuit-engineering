```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
alphas = np.linspace(0, 180, 1000)
v = 10/(2*np.pi)*(1+np.cos(alphas*np.pi/180))
```

```python
alphas, v
```

```python
plt.plot(alphas, v)
plt.grid()
plt.ylabel('V_load_mean, V')
plt.xlabel('alpha, degree')
```
